### Loading files ###

```python

f1 = open("harrypotter.txt", "r")
harry= f1.read()
harry
```

```python
'''
f2 = open("charlie.txt", "r")
charlie = f2.read()
charlie
'''

file17 = nltk.corpus.gutenberg.fileids()[16]
charlie = nltk.corpus.gutenberg.raw(file17)
charlie
```

```python
harry_len = len(harry)
charlie_len = len(charlie)

print("Length of book Harry Potter and the socerers stone= ",harry_len)
print("Length of book The tragedie of Macbeth = ",charlie_len)
```

### Tokenizing words ##

```python
harry_tokens = nltk.word_tokenize(harry)
harry_len = len(harry_tokens)
print(harry_len)
```

```python
charlie_tokens = nltk.word_tokenize(charlie)
charlie_len = len(charlie_tokens)
print(charlie_len)
```

### transforming words to lower case ###

```python
harry_words = [x.lower() for x in harry_tokens]
charlie_words = [x.lower() for x in charlie_tokens]
```

### creating frequency distributions ###

```python
from nltk import FreqDist
harry_fdist = FreqDist(harry_words)
harry_fdist_keys = list(harry_fdist.keys()) # Gives the unique words
len(harry_fdist_keys)
```

```python
 len(harry_fdist_keys) / harry_len
```

```python
charlie_fdist = FreqDist(charlie_words)
charlie_fdist_keys = list(charlie_fdist.keys()) # Gives the unique words
len(charlie_fdist_keys)
```

```python
len(charlie_fdist_keys) /charlie_len 
```

### most common words ###

```python
harry_topwords = harry_fdist.most_common(50)
harry_freq_percentage = [(word, (freq/harry_len)*100) for (word, freq) in harry_topwords]
harry_freq_percentage
```

```python
charlie_topwords = charlie_fdist.most_common(50)
charlie_freq_percentage = [(word, (freq/charlie_len)*100) for (word, freq) in charlie_topwords]
charlie_freq_percentage
```

# 1. Analyzing Harry Potter - Book1 : #

## 1.1  BiGram Analysis ##

### 1.1.1 BiGrams Using Raw Frequency: ###

```python
#simple bigrams:

from nltk.collocations import *
harry_bigram_measures = nltk.collocations.BigramAssocMeasures()
harry_finder_raw = BigramCollocationFinder.from_words(harry_words)
harry_scored_raw = harry_finder_raw.score_ngrams(harry_bigram_measures.raw_freq)

for bscore in harry_scored_raw[:50]:
    print(bscore)
```

```python
#Filtering data for better bigrams:

import re

#creating a function for non-alphabetic characters:

def alpha_filter(w):
    pattern = re.compile('^[^a-z]+$') # pattern for non-alphabetical characters
    if pattern.match(w):
        return True
    else:
        return False
# Getting stopwords from nltk
nltkstopwords = nltk.corpus.stopwords.words('english')
print(nltkstopwords[:20])

#adding more stopwords to the list
morestopwords = ['could','would','might','must','need','sha','wo','y',"'s","'d","'ll","'t","'m","'re","'ve", "n't"]
stopwords = nltkstopwords + morestopwords
    
# Applying filter to remove non-alphabetical tokens from bigram    
#harry_finder_raw.apply_word_filter(alpha_filter)

#Applying filter to remove stopwords

harry_finder_raw.apply_word_filter(alpha_filter)
harry_finder_raw.apply_word_filter(lambda w: w in stopwords)
harry_scored_raw_filtered = harry_finder_raw.score_ngrams(harry_bigram_measures.raw_freq)

for bscore in harry_scored_raw_filtered[:50]:
    print(bscore) #the values now make more sense
    
```

### 1.1.2 BiGrams Using PMI ###

```python
harry_finder_pmi = BigramCollocationFinder.from_words(harry_words)
harry_scored_pmi = harry_finder_pmi.score_ngrams(harry_bigram_measures.pmi)

for bscore in harry_scored_pmi[:50]:
    print(bscore)
```
